```python
import os
import uuid
import random
```

```python
province_ary = ["京", "津", "沪", "渝", "冀", "豫", "云", "辽", "黑", "湘",
                "皖", "鲁", "新", "苏", "浙", "赣", "鄂", "桂", "甘", "晋",
                "蒙", "陕", "吉", "闽", "贵", "粤", "青", "藏", "川", "宁",
                "琼"]
province_str = "".join(province_ary)
```

```python
alphabet_str = "abcdefghjklmnpqrstuvwxyz"
number_str = "0123456789012345678901234567890123456789"
all_str = alphabet_str + number_str
```

```python
records = []
total_num = 10000
for i in range(total_num):
    province = random.choice(province_str)
    first_let = random.choice(alphabet_str)
    random_str = ""
    for i in range(5):
        random_str += random.choice(all_str)
    records.append(province + first_let + random_str)
```

```python
records
```

```python
with open("plate2.txt", "w") as f:
    f.write("\n".join(records))
```
